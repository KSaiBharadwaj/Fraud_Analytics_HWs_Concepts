```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.gofplots import qqplot
from scipy.stats import shapiro
```

CSV File Loading

```python
data = pd.read_csv("/content/sample_data/Steel.csv")
column_name = "Error"
data_to_plot = data[column_name]
```

QQ Plot

```python
qqplot(data_to_plot, line='s')  # 's' for standardized line
plt.title('QQ Plot for ' + column_name)
plt.show()
```

The Sample data doesn't looks normal but to make sure we are conducting a Shapiro Wilk test on the dataset

```python
stat ,p = shapiro(data_to_plot)
print("Shapiro-Wilk Test: Shapiro:",stat," P-value: ",p)
```

As P-value is less than 0.05, the test concludes that the data is not normally-distributed
